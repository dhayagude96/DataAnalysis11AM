```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
dataset = [2,2,3,3,4,5,6,7,7,7,8,9,9,10,12,15,15,20,35]
```

```python
#Five Number Summary of Dataset:

print("Min value in Dataset : " , np.min(dataset))

#Q1 & Q3 

Q1,Q3 = np.percentile(dataset,[25,75])

print("Q1 : " , Q1)
print("Q3 : " , Q3)

#IQR (Q2) = Q3-Q1

Q2 = Q3-Q1

print("Q2 :" , Q2)

#LF & HF

LF = Q1 - 1.5*Q2
HF = Q3 + 1.5*Q2

print("LF : ",LF)
print("HF : ",HF)

print("Max : " , np.max(dataset))
```

```python
sns.boxplot(dataset)
```
