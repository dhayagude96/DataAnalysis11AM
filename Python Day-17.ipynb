# Encapsulation

Encapsulation is one of the fundamental concepts in object-oriented programming (OOP) and is supported in Python. It refers to the practice of bundling data (attributes or properties) and methods (functions) that operate on that data into a single unit called a class. This unit provides a way to control the access to the internal state of an object, meaning that the data is hidden from external access and can only be manipulated through well-defined methods.

```python
class emp:
    
    def __init__(self):
        #Access Modifier : private/public/defualt/protected
        self.__name = ""
        self.__pwd = ""
        self.__salary = 0
        self.empDetails()
        
        
    #get & Set method
    
    def get_name(self):
        return self.__name
    
    def set_name(self,new_name):
        temp1 = input("Enter Your Password")
        if temp1 == self.__pwd:
            self.__name = new_name
            
        else :
            print("Incorrect Password")
            
        
        
    def get_pwd(self):
        return self.__pwd
    
    def set_pwd(self,new_pwd):
        self.__pwd = new_pwd
        
    
        
        
    def empDetails(self):
        user_input = input("""
        Welcome to EduTECH
        1.Create Acc
        2.Login
        """)
        
        if user_input == "1":
            self.create_acc()
        else :
            self.login()
            
            
    def create_acc (self):
        self.__name = input("Enter Your Name :")
        self.__pwd = input("Set Your Password :")
        self.__salary = int(input("Enter Your Salary : "))
        print(f"Acc Created ..Hello {self.__name} welcome to EduTECH")
        
    
    def login(self):
        temp1 = input("Enter Your Name : ")
        temp2 = input("Enter Your Password")
        
        if temp1==self.__name and temp2 == self.__pwd :
            print(f"Hello {self.__name} Your current Salary is {self.__salary}")
            
        else:
            print("Plz check Your ID & Passowrd")
        
        
        
        
```

# Class Digram

```python
class info:
    
    def __init__(self):
        self.name = "Vishal"
        self.id = 101
        self.salary = 25000
        
    def _read1(self):
        print("Name : " , self.name)
        
    def read2(self):
        print("ID : ",self.id)
        
    def read3(self):
        print("Salary : ",self.salary)
```

```python
obj = info()

obj._read1()
obj.read2()
obj.read3()
```
