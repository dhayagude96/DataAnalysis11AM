# Exploratory Data Analysis

Exploratory Data Analysis (EDA) is a way of looking at and summarizing data to understand its main features. It involves using graphs and statistics to find patterns, trends, and potential issues in the data. EDA helps analysts and scientists get a quick overview of what the data looks like before diving into more detailed analysis or modeling.

# Key components of Exploratory Data Analysis include:



Descriptive Statistics:
---
    
    Calculating and presenting summary statistics, such as mean, median, mode, standard deviation, and percentiles, to describe the central tendency and variability of the data.

Data Visualization:
---
    
    Using various graphical representations, such as histograms, box plots, scatter plots, and heatmaps, to visually explore the distribution and relationships within the data.

Data Cleaning:
---
    
    Identifying and addressing issues in the data, such as missing values, outliers, and duplicates, to ensure the reliability of subsequent analyses.

Pattern Recognition:
---
    
    Detecting patterns, trends, and structures in the data that may lead to further insights or hypotheses about the underlying processes.

# df.info()

Provides a concise summary of a DataFrame, including the data types, non-null counts, and memory usage. Useful for understanding the overall structure of your dataset.

# df.describe()

Generates descriptive statistics of the central tendency, dispersion, and shape of the distribution of a DataFrame. Helpful for understanding the basic statistics of your numerical columns.

# df.shape

Returns a tuple representing the dimensions of the DataFrame (rows, columns). Useful for quickly checking the size of your dataset.

```python
print("Number of Rows & Columns in Data Frame : " , df.shape)
```

```python
print("Number of Rows in IPL : " , df.shape[0])
print("Number of Columns in IPL : " , df.shape[1])
```

```python
print("Number of Player Played IPL 2022 : " , len(df))
```

# df.groupby

Allows you to group data based on a column and perform aggregate functions (like sum, mean, count) on other columns within each group.

```python
nations = df.groupby('National Side')

nations.size()
```

```python
nations.size().plot(kind="bar")
```

```python
player_type = df.groupby('Type')

player_type.size()
```

```python
player_type.size().plot(kind="pie")
```
