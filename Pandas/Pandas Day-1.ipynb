# What is Pandas?

Pandas which is a popular open-source library for data manipulation and analysis. Pandas provides data structures and functions that make it easy to work with structured data, such as tabular data (like spreadsheets) or time series data.

# Data structures in Pandas are

Series:

A one-dimensional labeled array that can hold data of any type.

DataFrame: 

A two-dimensional table with rows and columns, similar to a spreadsheet or a SQL table.



Pandas allows you to read and write data from various file formats, including CSV, Excel, SQL databases, and more. It also provides powerful tools for data cleaning, filtering, grouping, aggregation, and visualization.

```python
import pandas as pd
```

```python
#Series:

#A one-dimensional labeled array that can hold data of any type.

project = pd.Series([5,2,10],index=["Ajinkya","Govind","Nitin"])

print(type(project))
print(project)
```

```python
#DataFrame:

#A two-dimensional table with rows and columns, similar to a spreadsheet or a SQL table.

#Using Dict
student_data = {
    "ID":[1,2,3],
    "Name":["Ajay","Nitin","Pooja"],
    "City":["Latur","Pune","Baramati"]
    
    
}

df = pd.DataFrame(student_data)

print(df)
```

```python
#Using Tuple

student_data = [
    (1,"Ajay","Latur"),
    (2,"Nitin","Pune"),
    (3,"Pooja","Baramati")
    
    
]

df1 = pd.DataFrame(data=student_data,columns=["ID","Name","City"])

df1
```

```python
#using List of Dict

student_data = [
    {"ID":1,"Name":"Ajay","City":"latur"},
    {"ID":2,"Name":"Nitin","City":"Pune"},
    {"ID":3,"Name":"Pooja","City":"Baramati"}
    
    
]
df2=pd.DataFrame(data=student_data,columns=["ID","Name","City"])

df2
```

```python
print("data Type of Data Frame : " , type(df2))

print("data Type of Name Col : " , type(df2["Name"]))
```
