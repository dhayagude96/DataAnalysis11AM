# Abstraction

Abstraction in programming is a fundamental concept that involves simplifying complex systems by breaking them into smaller, more manageable parts while hiding unnecessary details. It allows programmers to work with high-level, generalized representations of objects and processes, rather than getting bogged down in the specifics of their implementation. Abstraction makes code more readable, maintainable, and reusable.

# Abstraction is essential in programming for several important reasons:

Simplifies Complex Systems

Hides Implementation Details

Promotes Reusability

Facilitates Collaboration

Enhances Maintenance

Improves Code Understandability

Enforces a Clear Separation of Concerns

Adapts to Changing Requirements

```python
class RBI(ABC):
    
    #Interface :---> only have non-implemeted method
    
    @abstractmethod
    def rateOfInt(self):
        pass
    @abstractmethod
    def withdraw(self):
        pass
    @abstractmethod
    def deposit(self):
        pass
    
    
class Axis(RBI):
    
    
    def CustInfo(self):
        print("Welcome to axis Bank")
    
    def rateOfInt(self):
        print("Axis Bank have Home Loan Rate of Int 8%")
        
    def withdraw(self):
        print("Axis Bank have withdarw Limit upto 50k Per Day")
        
        
    def deposit(self):
        print("Axis Bank have Deposit Limit upto 90k Per Day")
        

        
class HDFC(RBI):
    
    
    def CustInfo(self):
        print("Welcome to HDFC Bank")
    
    def rateOfInt(self):
        print("HDFC Bank have Home Loan Rate of Int 11%")
        
    def withdraw(self):
        print("HDFC Bank have withdarw Limit upto 45k Per Day")
        
        
    def deposit(self):
        print("HDFC Bank have Deposit Limit upto 78k Per Day")
        
```

```python
a1 = Axis()

a1.CustInfo()
a1.deposit()
a1.rateOfInt()
a1.withdraw()
```

```python
h1 = HDFC()

h1.CustInfo()
h1.deposit()
h1.rateOfInt()
h1.withdraw()
```

```python
class cal(ABC):
    
    def __init__(self):
        self.num1 = 0
        self.num2 = 0
        
    def add(self):
        print("Add of Two Number is : " , self.num1+self.num2)
        
    def sub(self):
        print("Sub of Two Number is : " , self.num1-self.num2)
        
    def div(self):
        print("Div of Two Number is : " , self.num1/self.num2)
        
    @abstractmethod   
    def UserIn(self):
        pass
    
    
class Android(cal):
    
    def __init__(self):
        super().__init__()
        
        
        
    def UserIn(self):
        print("---ANDROID---")
        self.num1 = int(input("Enter Number :"))
        self.num2 = int(input("Enter Number :"))
        
    
    def __init__(self):
        super().__init__()
        
        
        
    def UserIn(self):
        print("---ANDROID---")
        self.num1 = int(input("Enter Number :"))
        self.num2 = int(input("Enter Number :"))
        
        
class IOS(cal):
    
    def __init__(self):
        super().__init__()
        
        
        
    def UserIn(self):
        print("---IOS---")
        self.num1 = int(input("Enter Your 1st Number :"))
        self.num2 = int(input("Enter Your 2nd Number :"))
               
```
