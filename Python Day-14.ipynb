# What is filter Function in Python?

In Python, filter() is a built-in function that is used to filter elements from an iterable (such as a list, tuple, or other iterable objects) based on a specified condition. It creates a new iterable containing only the elements that satisfy the given condition.

```python
L1 = list(range(5,500))

#list(filter(lambda L1: L1%2==0 , L1))

list(filter(lambda x : (x>=100 and x<=150),L1))
```

# What is Reduce Function in Python

The reduce function in Python is part of the functools module and is used for performing a repetitive operation on the elements of an iterable (e.g., a list) to reduce it to a single accumulated result. It is especially useful when you need to apply a binary function (a function that takes two arguments) cumulatively to the items of a sequence.

The reduce function takes two arguments:

A binary function (a function that takes two arguments).

An iterable (e.g., a list) of values to be reduced.

```python
from functools import reduce

x = list(range(1,500))#1,2,3,4,5


reduce(lambda x,y : x+y , x)
```

```python
#Regular Function

def regFun():
    return 10

def Gen():
    yield 10
    yield 20
    yield 30
```

```python
regFun()

g = Gen()

print(next(g))
print(next(g))
print(next(g))
```

```python
for data in g :
    print(data)
```

```python
def Numcal(x):
    for i in x :
        db = i*2
        sq = i*i
        cube = i * i *i
        ev = "Even"if i%2==0 else "odd"
        #return db,sq,cube,ev
        yield db,sq,cube,ev
```

```python
L1 = list(range(1,1000))

for data in Numcal(L1):
    print(data)
```
