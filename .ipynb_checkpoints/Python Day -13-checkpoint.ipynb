# What is Lambda Function?

In Python, a lambda function is a small, anonymous (unnamed) function defined using the lambda keyword. Lambda functions are also sometimes called anonymous functions or lambda expressions. They are used for creating simple, one-line functions on the fly without having to define a formal function using the def keyword.


The general syntax of a lambda function is:

lambda arguments: expression

```python
#Noramal Function

def double(n):
    x = n*2
    return x

x=lambda n : n*2

x(10)
```

# Differnace Between Lambda & Regular Function

```python
#Write a program to return Cube of enter Value

y = lambda a : a*a*a

print("Cube of 5 : " , y(5))

#Write a program to caluclate avg of three Number

x = lambda a,b,c : (a+b+c)/3

print("Avg of three Number is : " , x(23,24,21))

#Write a program to calculate dobule/cube/square of entered value

z = lambda x : (x*2 , x*x , x*x*x)

print("Return dobule/square/cube of 3 : "  , z(3))

#Write a program to print number is odd or even

b = lambda x : "EVEEN" if x%2 == 0 else "odd"

print("Check number enter is odd or even : " , b(25))
```

# What is Map in Python?


In Python, map is a built-in function used for applying a given function to each item in an iterable (such as a list, tuple, or other iterable objects) and returning an iterator of the results. The map function takes two or more arguments:

A function: This is the function that you want to apply to each item in the iterable.

One or more iterable(s): These are the sequences of data that you want to apply the function to. You can provide multiple iterables if the function takes multiple arguments.

```python
x = [12,25,13,10,15,18,10]

#y = lambda a : a*2

list(map(lambda a : a*2 ,x))

y = list(range(10,60))

#list(map(lambda x : (x*2,x*x,x*x*x) , y))

list(map(lambda x : ( x > 15 and x < 40) , y))
```

# What is filter Function in Python?

In Python, filter() is a built-in function that is used to filter elements from an iterable (such as a list, tuple, or other iterable objects) based on a specified condition. It creates a new iterable containing only the elements that satisfy the given condition.

```python
y = list(range(10,50))

#list(filter(lambda x : x%2==0 , y ))

list(filter(lambda x : ( x > 15 and x < 40) , y))
```

# What is Reduce Function in Python

The reduce function in Python is part of the functools module and is used for performing a repetitive operation on the elements of an iterable (e.g., a list) to reduce it to a single accumulated result. It is especially useful when you need to apply a binary function (a function that takes two arguments) cumulatively to the items of a sequence.

The reduce function takes two arguments:

A binary function (a function that takes two arguments).

An iterable (e.g., a list) of values to be reduced.

```python
from functools import reduce

x = list(range(1,5))

reduce(lambda x,y : x+y , x)
```
