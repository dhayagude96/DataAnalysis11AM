Polymorphism

Polymorphism is a key principle in OOP, and it promotes code reusability, flexibility, and modularity. It simplifies the writing of code that can work with a variety of objects without needing to know their specific types, making it a powerful tool for designing and implementing complex software systems

There are two main types of polymorphism

Compile-time (or static) polymorphism: 
    
    This is also known as method overloading or operator overloading. It occurs when multiple methods or operators with the same name are defined in a class, but they have different parameter lists. The correct method or operator to call is determined at compile time based on the arguments used in the function call or operation.


Runtime (or dynamic) polymorphism: 
    
    This is typically achieved through method overriding. It allows different classes to provide their own implementation for a method that is declared in a base class or interface. When a method is called on an object, the actual implementation that gets executed is determined at runtime based on the object's type, not at compile time.








```python
class PolyDemo:
    
    def add(self,a=0,b=0,c=0):
        print("Add is : " , a+b+c)
         
```

```python
a1 = PolyDemo()

a1.add()
a1.add(10,12)
a1.add(5,18,18)
a1.add(10.5,21.8)
a1.add(10.5,21.8,56.28)
```
