# what class & Object in OOP?

# Class:

A class is a blueprint or template for creating objects. It defines the structure and behavior that its objects will have.

It serves as a blueprint that specifies the attributes (also called fields or properties) and methods (functions) that objects created from the class will have.

Classes are typically used to represent real-world entities or abstract concepts in your software.

# Object:

An object is an instance of a class. It is a concrete realization of the class blueprint, with its own unique data and state.

Objects have attributes (properties) and can perform actions (methods) defined in their class.

You can create multiple objects from the same class, each with its own set of attribute values and the ability to call methods.
